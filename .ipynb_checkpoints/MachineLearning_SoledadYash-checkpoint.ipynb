## Data Cleaning & Transformation (Melting / Pivoting)

```python
print(df1.columns.tolist())
```

```python
df1.columns = df1.columns.str.replace(r"\s*\[YR\d+\]", "", regex=True)

selected_series = [
    "Control of Corruption: Estimate",
    "Government Effectiveness: Estimate",
    "Regulatory Quality: Estimate",
    "Rule of Law: Estimate"]

df1_filtered = df1[df1["Series Name"].isin(selected_series)]

df1_melt = df1_filtered.melt(
    id_vars=["Country Name", "Series Name"],
    var_name="Year",
    value_name="Value")


df1_melt["Year"] = pd.to_numeric(df1_melt["Year"], errors="coerce")
df1_melt["Value"] = pd.to_numeric(df1_melt["Value"], errors="coerce")


df1_clean = df1_melt.pivot_table(
    index=["Country Name", "Year"],
    columns="Series Name",
    values="Value"
).reset_index()

print("WGI clean:", df1_clean.shape)
df1_clean.head(10)
```

```python
print("Unique quantity:", df2["indicator"].nunique())
print(df2["indicator"].unique()[:50]) 
```

```python
# PIVOT HDR 

df2_sub = df2[["country", "year", "indicator", "value"]].copy()
df2_sub["year"]  = pd.to_numeric(df2_sub["year"], errors="coerce")
df2_sub["value"] = pd.to_numeric(df2_sub["value"], errors="coerce")
df2_sub = df2_sub[df2_sub["year"].between(2013, 2023, inclusive="both")]

df2_pivot = df2_sub.pivot_table(
    index=["country", "year"],
    columns="indicator",
    values="value",
    aggfunc="mean"
).reset_index()

selected_indicators = [
    "Human Development Index (value)",
    "Gender Inequality Index (value)",
    "Expected Years of Schooling (years)",
    "Mean Years of Schooling (years)",
    "Gross National Income Per Capita (2021 PPP$)",
    "Life Expectancy at Birth (years)",
    "Population with at least some secondary education, female (% ages 25 and older)",
    "Population with at least some secondary education, male (% ages 25 and older)",
    "Labour force participation rate, female (% ages 15 and older)",
    "Labour force participation rate, male (% ages 15 and older)"]

df2_pivot = df2_pivot[
    [c for c in df2_pivot.columns if c in ["country", "year"] + selected_indicators]]

merged = pd.merge(
    df1_clean,
    df2_pivot,
    left_on=["Country Name", "Year"],
    right_on=["country", "year"],
    how="inner")


wgi_cols = [
    "Control of Corruption: Estimate",
    "Government Effectiveness: Estimate",
    "Regulatory Quality: Estimate",
    "Rule of Law: Estimate"]

final_order = (
    ["Country Name", "Year"]
    + [col for col in wgi_cols if col in merged.columns]
    + [col for col in selected_indicators if col in merged.columns])

df = merged.drop(columns=[c for c in ["country", "year"] if c in merged.columns])
df = df.loc[:, final_order].sort_values(["Country Name", "Year"]).reset_index(drop=True)
df.head(10)
```

```python
features = ['corruption', 'gov_effectiveness', 'regulatory_quality', 'rule_of_law', 'hdi', 'gender_inequality', 'expected_school', 'mean_school','gni_per_capita', 'life_expectancy']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[features])
scaled_df = pd.DataFrame(X_scaled, columns=features)
```

```python
# Detect outliers using the IQR (Interquartile Range) method
Q1 = scaled_df.quantile(0.25)
Q3 = scaled_df.quantile(0.75)
IQR = Q3 - Q1

# Count how many outliers each variable has
outlier_counts = ((scaled_df < (Q1 - 1.5 * IQR)) | (scaled_df > (Q3 + 1.5 * IQR))).sum()
print("Outliers per variable:")
print(outlier_counts)
```

```python
# Clip only GNI extremes to reduce dominance (keeps all rows)
scaled_df['gni_per_capita'] = np.clip(
    scaled_df['gni_per_capita'],
    scaled_df['gni_per_capita'].quantile(0.01),
    scaled_df['gni_per_capita'].quantile(0.99))
```

# Clustering Analysis

```python
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import silhouette_score, davies_bouldin_score

features = ['corruption', 'gov_effectiveness', 'regulatory_quality', 'rule_of_law',
            'hdi', 'gender_inequality', 'expected_school', 'mean_school',
            'gni_per_capita', 'life_expectancy']
X = scaled_df[features].copy()
silhouette_scores = []
davies_scores = []
K_range = range(2, 8)
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42) labels = kmeans.fit_predict(X) silhouette_scores.append(silhouette_score(X, labels))
    davies_scores.append(davies_bouldin_score(X, labels))


fig, ax1 = plt.subplots(1, 2, figsize=(12, 5))
sns.lineplot(x=list(K_range), y=silhouette_scores, marker='o', ax=ax1[0])
ax1[0].set_title("Silhouette Score by K", weight='bold')
ax1[0].set_xlabel("Number of Clusters (k)")
ax1[0].set_ylabel("Silhouette Score")
sns.lineplot(x=list(K_range), y=davies_scores, marker='o', ax=ax1[1])
ax1[1].set_title("Davies–Bouldin Index by K", weight='bold')
ax1[1].set_xlabel("Number of Clusters (k)")
ax1[1].set_ylabel("Davies–Bouldin Index")
plt.tight_layout()
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
X['cluster_kmeans'] = kmeans.fit_predict(X)
agg = AgglomerativeClustering(n_clusters=3)
X['cluster_agg'] = agg.fit_predict(X)
sil_k = silhouette_score(X[features], X['cluster_kmeans'])
db_k = davies_bouldin_score(X[features], X['cluster_kmeans'])
sil_a = silhouette_score(X[features], X['cluster_agg'])
db_a = davies_bouldin_score(X[features], X['cluster_agg'])

print(f"K-Means, Silhouette: {sil_k:.3f} | Davies–Bouldin: {db_k:.3f}")
print(f"Agglomerative, Silhouette: {sil_a:.3f} | Davies–Bouldin: {db_a:.3f}")
```

```python
sns.set(style="whitegrid", font_scale=1.1)
plt.figure(figsize=(10,6))
sns.scatterplot(data=X, x='gov_effectiveness', y='hdi',
                hue='cluster_kmeans', palette='Set2', s=80)
plt.title("Country Clusters by Governance Effectiveness and HDI (K-Means)", weight='bold')
plt.xlabel("Government Effectiveness (z-score)")
plt.ylabel("Human Development Index (HDI)")
plt.legend(title="Cluster ID")
plt.tight_layout()
plt.show()

plt.figure(figsize=(10,6))
sns.scatterplot(data=X, x='gov_effectiveness', y='hdi',
                hue='cluster_agg', palette='Pastel1', s=80)
plt.title("Country Clusters by Governance Effectiveness and HDI (Agglomerative)", weight='bold')
plt.xlabel("Government Effectiveness (z-score)")
plt.ylabel("Human Development Index (HDI)")
plt.legend(title="Cluster ID")
plt.tight_layout()
plt.show()
```

```python
cluster_summary = X.groupby('cluster_kmeans')[features].mean().round(3)
print("\nCluster Summary (K-Means):")
display(cluster_summary)
```

# Institutional Incongruence & Governance–Development Coherence

```python
from scipy.stats import zscore
df['governance_index'] = df[['corruption','gov_effectiveness','regulatory_quality','rule_of_law']].mean(axis=1)
country_mean = (df.groupby('country', as_index=False)[['hdi','governance_index','gender_inequality']].mean())
country_mean['hdi_z'] = zscore(country_mean['hdi'])
country_mean['gov_z'] = zscore(country_mean['governance_index'])
country_mean['incongruence'] = abs(country_mean['hdi_z'] - country_mean['gov_z'])
outliers = country_mean.sort_values('incongruence', ascending=False).head(10)
sns.set(style="whitegrid", font_scale=1.1)
plt.figure(figsize=(13,8))
scatter = sns.scatterplot(data=country_mean, x='governance_index', y='hdi', hue='gender_inequality', size='incongruence',
    sizes=(40,400), palette='coolwarm_r',  alpha=0.85, edgecolor='white', linewidth=0.6)


plt.title("Governance–Development Coherence by Gender Inequality (2013–2023 Average)", fontsize=15, weight='normal', pad= 3, loc= "center")
plt.suptitle( "Each circle = one country | X: Governance Quality | Y: Human Development |""Color: Gender Inequality | Size: Institutional Incongruence",fontsize=10, y=0.99
        )
plt.xlabel("Governance Index (avg of corruption control, government effectiveness, regulatory quality, rule of law)", fontsize=10)
plt.ylabel("Human Development Index (HDI)", fontsize=10)
plt.axvline(0, color='gray', linestyle='--', alpha=0.4)
plt.axhline(0.7, color='gray', linestyle='--', alpha=0.4)
plt.text(1.4, 0.94, "High Governance / High HDI\n(Coherent Systems)", color='black', fontsize=9, ha='center')
plt.text(-1.6, 0.4, "Low Governance / Low HDI\n(Fragile States)", color='black', fontsize=9, ha='center')
plt.text(0, 0.55, "Incongruent Zone\nGovernance ≠ Development", color='black', fontsize=9, ha='center')
handles, labels = scatter.get_legend_handles_labels()
plt.legend(
    handles=handles[1:], labels=labels[1:],
    title="Legend:\nGender Inequality (color)\nGovernance–HDI Incongruence (size)",
    bbox_to_anchor=(1.02, 1), loc='upper left', frameon=True)
plt.tight_layout()
plt.show()
```

# ARIMA and SARIMA for times series

```python
plt.figure(figsize=(10,6))
sns.lineplot(data=hdi_trend, x='year', y='hdi', marker='o', color='steelblue')
plt.title("Global Human Development Index (HDI) Trend, 2013–2023", weight='bold')
plt.xlabel("Year")
plt.ylabel("Average HDI")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

print(hdi_trend.tail())
```

```python
from pmdarima import auto_arima

hdi_ts = hdi_trend.set_index('year')['hdi']
model = auto_arima(
    hdi_ts,
    seasonal=False,
    stepwise=True,
    suppress_warnings=True)
n_periods = 10
forecast, conf_int = model.predict(n_periods=n_periods, return_conf_int=True)
future_years = range(hdi_ts.index[-1] + 1, hdi_ts.index[-1] + n_periods + 1)
forecast_df = pd.DataFrame({
    'year': future_years,
    'hdi_forecast': forecast,
    'lower_ci': conf_int[:, 0],
    'upper_ci': conf_int[:, 1]})
combined_10y = pd.concat([
    hdi_trend.assign(type='Historical'),
    forecast_df.assign(type='Forecast')])
sns.set(style="whitegrid", font_scale=1.1)
plt.figure(figsize=(11,6))
sns.lineplot(
    data=combined_10y[combined_10y['type']=="Historical"],
    x='year', y='hdi', marker='o', linewidth=2.2,
    color='steelblue', label='Historical')
sns.lineplot(
    data=combined_10y[combined_10y['type']=="Forecast"],
    x='year', y='hdi_forecast', marker='o', linewidth=2.2,
    linestyle='--', color='tomato', label='Forecast')
plt.fill_between(forecast_df['year'], forecast_df['lower_ci'], forecast_df['upper_ci'],
                 color='orange', alpha=0.2, label='95% Confidence Interval')

plt.title("Global HDI Forecast (ARIMA, 2013–2033)", weight='bold')
plt.xlabel("Year")
plt.ylabel("Average Human Development Index (HDI)")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
